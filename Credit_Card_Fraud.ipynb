```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
```

```python
data = pd.read_csv('creditcard.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.drop(columns=['Time'], inplace=True)
```

```python
data.Class.value_counts()
```

```python
legal = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]
```

```python
legal.shape
```

```python
fraud.shape
```

```python
data.Class.value_counts()
```

**Under Sampling:** We are going to reduce the count of legal

```python
fraud.shape
```

```python
legal_sample = data.sample(n=473)
```

We need to merge Legal and Fraud transations and make it as one complete dataset.

```python
data_undersampling = pd.concat([legal_sample, fraud], ignore_index=True)
```

```python
data_undersampling.shape
```

```python
data_undersampling.Class.value_counts()
```

```python
x= data_undersampling.drop(columns=['Class'])
```

```python
y= data_undersampling['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
predict = model.predict(x_test_scaled)
```

```python
predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(predict, y_test)
```

**OverSampling**

```python
data.Class.value_counts()
```

```python
sm = SMOTE(random_state=42)
```

```python
x = data.drop(columns=['Class'])
y = data['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
x_oversaample, y_oversample = sm.fit_resample(x, y.ravel())
```

```python
x_oversaample.shape
```

```python
y_oversample.shape
```

```python
pd.value_counts(y_oversample)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_oversaample, y_oversample, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
x_train_scale = scaler.fit_transform(X_train)
y_train_scale = scaler.fit_transform(X_test)
```

```python
model2 = LogisticRegression()
```

```python
model2.fit(x_train_scale, y_train)
```

```python
pred = model2.predict(y_train_scale)
```

```python
accuracy_score(pred, y_test)
```
